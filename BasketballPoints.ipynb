##### Question 1: Is there a correlation between turnovers and points per game?

```python
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import requests
import seaborn as sns
sns.set_style("darkgrid")
```

```python
url = 'https://www.espn.com/nba/stats/team/_/view/team/table/offensive/sort/avgPoints/dir/desc'
results = pd.read_html(url)
Data = pd.DataFrame(results[1])
Teams = pd.DataFrame(results[0])
Teams = Teams.drop(['RK'], axis=1)
Data.index = Teams
Data
sns.set(rc = {'figure.figsize':(15,8)})
```

### Question 1: Do Point per game correlate with the amount of 3 pointers made?

```python
graph = sns.scatterplot(
                        x='PTS', #Points per game
                        y='3PM', #3 pointers made
                        hue= Data.index,
                        data=Data 
                        )
graph.get_legend().set_bbox_to_anchor((1, 1)) #Fixes the data so it shows it with the teams on the right
graph
plt.savefig('scatter.png')
```

##### There is a correlation between teams that make 3 pointers versus the points per game. It's a positive correlation of 3 pointers made by teams and Points made by game. there is some obvious small outliers in the graph but the graph shows a positive correlation.

### Question 2: Is there a correlation between field goal percentage and 3-point attempts? 

```python
graph = sns.scatterplot(
                        x='FG%', #Field goals percentage
                        y='3PA', #3 pointers attempted
                        hue= Data.index,
                        data=Data 
                        )
graph.get_legend().set_bbox_to_anchor((1, 1)) #Fixes the data so it shows it with the teams on the right
graph
plt.savefig('scatter.png')
```

### There is minimal to no correlation between the amount of 3 points attempter versus the field goal percentage. There seems to be a very small not noticable positive correlation but not enough to classify as a correlation. 

### Question 3: Which of these factors correlates strongly to Points per game(PPG)?

##### The factors that correlates strongly to points Per game would be the 3-points made throughout the NBA season. As shown with the graph of the 3 pointers made, the amount of points made based on 3 pointers define the points per game.
